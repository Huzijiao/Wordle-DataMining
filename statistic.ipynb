```python
import pandas as pd
import operator
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
fig = plt.figure()
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号
```

```python
data = pd.read_excel('Problem_C_Data_Wordle_pre.xlsx') # 默认读取第一个sheet
print(data.head()) # 预览数据
```

```python
data1 = data.iloc[1:,[3]].values
print(len(data1))
```

```python
Most_Fre = []
Least_Fre = []
Class = []
Most_Combi = []
```

```python
#先进行最常见字母占比考量，分别e,a,r,o,t,以及总和
for word in data1:
    w = word[0]
    e1 = 0
    a1 = 0
    r1 = 0
    o1 = 0 
    t1 = 0
    for i in w:
        if(i=='e'):
            e1 = e1 + 1
        if(i=='a'):
            a1 = a1 + 1
        if(i=='r'):
            r1 = r1 + 1
        if(i=='o'):
            o1 = o1 + 1
        if(i=='t'):
            t1 = t1 + 1
    t = [w,e1,a1,r1,o1,t1,e1+a1+r1+o1+t1]
    Most_Fre.append(t)
aa = np.array(Most_Fre)
df = pd.DataFrame(aa)
df.to_csv("Most_Fre.csv")
```

```python
#进行最少见字母占比考量，分别j,q,z,x,v,以及总和
for word in data1:
    w = word[0]
    j1 = 0
    q1 = 0
    z1 = 0
    x1 = 0 
    v1 = 0
    for i in w:
        if(i=='j'):
            j1 = j1 + 1
        if(i=='q'):
            q1 = q1 + 1
        if(i=='z'):
            z1 = z1 + 1
        if(i=='x'):
            x1 = x1 + 1
        if(i=='v'):
            v1 = v1 + 1
    t = [w,j1,q1,z1,x1,v1,j1+q1+z1+x1+v1]
    Least_Fre.append(t)
aa = np.array(Least_Fre)
df = pd.DataFrame(aa)
print(len(df))
df.to_csv("Least_Fre.csv")
```

```python
#进行词汇种类统计
for word in data1:
    w = word[0]
    a = [0]*256
    for i in w:
        a[ord(i)] = 1
    SUM = 0
    for i in range(0,200):
        SUM = SUM + a[i]
    t = [w,SUM]
    Class.append(t)
aa = np.array(Class)
df = pd.DataFrame(aa)
df.to_csv("Class.csv")
```

```python
#最后统计表内组合数，分别为er,ar,or,ea,ri,ir,ie,ee,ur,th,ck,dy,pt
Com = ["er","ar","or","ea","ri","ir","ie","ee","ur","th","ck","dy","pt"]
for word in data1:
    w = word[0]
    o = [w]
    for j in Com:
        SUM = 0
        if j in w:
            SUM = 1
        o.append(SUM)
    Most_Combi.append(o)
aa = np.array(Most_Combi)
df = pd.DataFrame(aa)
df.to_csv("Most_Combi.csv")
```
